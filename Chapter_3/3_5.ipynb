### 結合処理/集計処理

```python
# {コード3.28 3.5節の準備}
df_sale1_1 = pd.read_csv('./in/4_company_sales_data1.csv')
df_sale1_2 = pd.read_csv('./in/4_company_sales_data2.csv')
df_info = pd.read_csv('./in/4_company_attri_data.csv')

print('df_sale1_1'); display(df_sale1_1.head())
print(len(df_sale1_1))

print('df_sale1_2'); display(df_sale1_2.head())
print(len(df_sale1_2))

print('df_info'); display(df_info.head())
print(len(df_info))
```

```python
# {コード3.29 縦結合}
df_sale2_1= pd.concat([df_sale1_1, df_sale1_2],axis=0)   # 縦の結合

print('df_sale2_1'); display(df_sale2_1.head())
print(len(df_sale2_1))
```

```python
# {コード3.30 縦結合後の推奨処理}
#並び替え
df_sale2_2 = df_sale2_1.sort_values(['企業管理コード','決算期年'],ascending=[True,True])
# レコードの重複削除
df_sale2_3 = df_sale2_2[~df_sale2_2[['企業管理コード','決算期年']].duplicated()]    # 重複行の抽出

print('df_sale2_3'); display(df_sale2_3.head())
print(len(df_sale2_3))
```

```python
# {コード3.31 リストを用いた一括読み込み処理}
all_csv_files2 = [file for file in glob(os.path.join('./in/*_sales_*.csv'))]

df1_ap=[]
for file in all_csv_files2:
    df1 = pd.read_csv(file) 
    df1_ap.append(df1)

df_sale2_1 = pd.concat(df1_ap, axis=0, sort=False)

print('df_sale2_1'); display(df_sale2_1.head())
print(len(df_sale2_1))
```

```python
# {コード3.32 データの横結合}
df_mer1=pd.merge( df_info, df_sale2_3,  on='企業管理コード',how='inner')

print('df_mer1'); display(df_mer1.head())
print(len(df_mer1))
```

```python
# {コード3.33 売上高と利益の縦構造化}
df_mer1_1=pd.melt(df_mer1, id_vars=['企業管理コード','決算期年'], value_vars=['売上高_百万円','利益_千円'], value_name='Value')

print('df_mer1_1'); display(df_mer1_1.head())
print(len(df_mer1_1))
```

```python
# {コード3.34 区分別の売上高の平均}
df_mer2 =df_mer1[['決算期年','所在地_都道府県','売上高_百万円']].groupby(['決算期年','所在地_都道府県']).mean()

print('df_mer2'); display(df_mer2.head())
print(len(df_mer2))
```

```python
# {コード3.35 データの生成}
df=pd.DataFrame({'単語':pd.Categorical(["激安","激安","限定","高速","学習","高速","AI","学習","AI","AI"]),
                 '文章':pd.Categorical(["D1","D1","D1","D1","D2","D2","D2","D2","D2","D2"])})

display(df.head(10))
```

```python
# {コード3.36 文章と単語のクロス集計}
crs=pd.crosstab(df['文章'],df['単語'])

display(crs.head(10))
```

```python
# {コード3.37 クロス集計の小計と規格化}

# 小計をAll として出力
crs2=pd.crosstab(df['文章'],df['単語'],margins=True)

# 全体が1になるように規格化
crs3=pd.crosstab(df['文章'],df['単語'],margins=True,normalize='all')

# 行合計が1になるように規格化
crs4=pd.crosstab(df['文章'],df['単語'],margins=True,normalize='index')

# 列合計が1になるように規格化
crs5=pd.crosstab(df['文章'],df['単語'],margins=True,normalize='columns')


display(crs2.head(10))
display(crs3.head(10))
display(crs4.head(10))
display(crs5.head(10))
```
