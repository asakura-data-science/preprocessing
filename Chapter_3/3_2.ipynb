```python
# {コード3.4 3.2節の準備}
df_col1 = pd.read_csv('./in/1_company_data.csv',dtype = {'所在地_市区郡コード':'object'}) 

print('df_col1'); display(df_col1.head())
print(len(df_col1))
```
