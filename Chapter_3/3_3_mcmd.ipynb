### 前準備

```python
#{コード3.1 3章の準備}
import os
import nysol.mcmd as nm

os.makedirs('out_m',exist_ok=True)
```

### 値操作

```python
# {コード3.12 3.3節の準備}
#   -> nysol_pythonの場合CSVファイルのまま処理できるため、前準備は特に不要
file_val1='./in/2_company_data.csv'
display(nm.p(file_val1, head=5, tail=1))
```

```python
# {コード3.13 文字列の切り出し処理}
file_val2='./out_m/df_val2.csv'
#項目値の文字列演算にはnm.mcal()メソッドを使用する.
#  c=パラメータの演算指定にて, 先頭からの切り出しはleft()関数, 末尾からの切り出しはright()関数, 部分切り出しはmid()関数を用いる.
f=None
f<<=nm.mcal(i=file_val1, c='left($s{所在地_市区郡コード},2)', a='所在地_都道府県コード', o=file_val2)
f.run(msg="on")

display(file_val2, nm.p(file_val2, head=5, tail=1))
```

```python
# {コード3.14 コードと名前の結合}
file_val3='./out_m/df_val3.csv'
#項目値の結合にはnm.mcal()メソッドを使用する.
#  c=パラメータの演算指定にて,cat()関数を用いる. cat()関数は第1引数に区切り文字, 第2引数以降に結合項目を指定する.
#  区切り文字が不要な場合は""（空文字）を指定すればよい.
f=None
f<<=nm.mcal(i=file_val2, c='cat("_", $s{所在地_都道府県コード}, $s{所在地_都道府県名})', a='所在地_都道府県', o=file_val3)
f.run(msg="on")

display(file_val3, nm.p(file_val3, head=5, tail=1))
```

```python
# {コード3.15 文字列の置換}
file_val4='./out_m/df_val4.csv'
#文字列の置換にはnm.mchgstr()メソッドを使用する.
#   f=パラメータに置換対象の項目を指定する.
#   c=パラメータに 置換対象文字列:置換後文字列 を指定する（複数指定する場合はカンマ区切りの1つの文字列を指定する）. 
#   置換後の新しい項目を作成する場合はA=Trueオプションを指定し, f=パラメータにて新しい項目名を指定する.
#   F=Trueオプションを指定することで、c=パラメータで指定した置換ルールが存在しなかった場合、元の値がそのまま出力される.
f=None
f<<=nm.mchgstr(i=file_val3, f='従業員数:従業員数_int', A=True, F=True, c='-:0', o=file_val4)
f.run(msg="on")

display(file_val4, nm.p(file_val4, head=5, tail=1))
```

```python
# {コード3.16 生産性指標の算出}
file_val5='./out_m/df_val5.csv'
#数値演算はnm.mcal()メソッドを使用する.
#   c=パラメータ内で使用するround()関数の第2パラメータで指定する基数の詳細はマニュアルを参照
#   小数点以下二桁目を四捨五入する場合は基数に0.0001を指定する.
f=None
f<<=nm.mcal(i=file_val4, c='round(${売上高_百万円}/${従業員数_int},0.0001)', a='生産性指標', o=file_val5)
f.run(msg="on")

display(file_val5, nm.p(file_val5, head=5, tail=1))
```

```python
#{3.2 補足}
# 売上高_百万円}/従業員数_int の計算で 従業員数_int の値が0の場合, 計算結果値は空文字（null）になる.
#   nm.mdelnull()メソッドはf=パラメータで指定した項目値が空文字（null）の行を削除するが,　
#   r=Trueオプションを指定することで条件反転となり該当行を抽出することができる.
nm.mdelnull(i=file_val5, f='生産性指標', r=True)
```
