### 値操作

```python
# {コード3.12 3.3節の準備}
df_val1 = pd.read_csv('./in/2_company_data.csv',dtype = {'所在地_市区郡コード':'object'}) 

print('df_val1'); display(df_val1.head())
print(len(df_val1))
```

```python
# {コード3.13 文字列の切り出し処理}
df_val2=df_val1.assign(所在地_都道府県コード=df_val1['所在地_市区郡コード'].str[0:2])

print('df_val2'); display(df_val2.head())
print(len(df_val2))
```

```python
# {コード3.14 コードと名前の結合}
df_val3=df_val2.assign(所在地_都道府県=df_val2['所在地_都道府県コード'] + '_' +df_val2['所在地_都道府県名'])

print('df_val3'); display(df_val3.head())
print(len(df_val3))
```

```python
# {コード3.15 文字列の置換}
df_val4 = df_val3.copy()

df_val4['従業員数_int'] = df_val4['従業員数'].replace('-', '0').astype(np.int64)

df_val4['従業員数_int'].dtype
```

```python
# {コード3.16 生産性指標の算出}
df_val5=df_val4.assign(生産性指標=round( df_val4['売上高_百万円'] / (df_val4['従業員数_int']),4))

print('df_val5'); display(df_val5.head())
print(len(df_val5))
```

```python
#{3.2 補足}
# inf は infinity の略で、正の無限大 (+∞) を意味する特殊な浮動小数点数型データです。
df_val5[df_val5['従業員数_int'] == 0]
```
